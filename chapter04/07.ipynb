# データの可視化

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 折れ線グラフ

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 7, 4, 8])

plt.plot(x, y, color='r')
```

## 散布図

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 7, 4, 8])

plt.scatter(x, y, color='r')
```

## 曲線のグラフ

```python
# y = x^2

x = np.linspace(-5, 5, 300)
y = x**2

plt.plot(x, y, color='r')
```

## 複数の線を表示する

```python
x = np.linspace(-5, 5, 300)
y1 = x**2
y2 = (x - 2)**2

plt.plot(x, y1, color='r')
plt.plot(x, y2, color='k', linestyle='--')
```

## ヒストグラム

```python
# サイコロを１０回振って目の合計を計算する事を１０００回やったときの目の合計の分布がどうなるかをヒストグラムで表してみる。

np.random.seed(0)

l = []
for _ in range(1000):
    l.append(np.random.randint(1, 7, size=10).sum())
    
plt.hist(l, 20, color='gray')
```

```python
# 区間を明示的に指定する事もできる。
plt.hist(l, np.arange(15, 55, 2))
```

## 複数のグラフを並べて表示する

```python
x = np.linspace(-5, 5, 300)
sin_x = np.sin(x)
cos_x = np.cos(x)

fig, axes = plt.subplots(2, 1)
axes[0].set_ylim([-1.5, 1.5])
axes[1].set_ylim([-1.5, 1.5])
axes[0].plot(x, sin_x, color='r')
axes[1].plot(x, cos_x, color='k')
```

```python
x = np.linspace(-5, 5, 300)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y, color='r')
```

## 等高線の描画

```python
# x^2 + (y^2)/4 = k
# を満たす曲線を描画する。これはkを固定すると楕円になる。k=1,2,3,4,5のときの曲線を描いてみる。

def f(x, y):
    return x**2 + y**2 / 4

x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)

xmesh, ymesh = np.meshgrid(x, y)
z = f(xmesh.ravel(), ymesh.ravel()).reshape(xmesh.shape)
#.ravel()で平坦化しf（）に入れて返ってきたものをxmeshの形に整形する。

plt.contour(x, y, z, color='k', levels=[1, 2, 3, 4, 5])
```

```python
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
xmesh, ymesh = np.meshgrid(x, y)
z = f(xmesh, ymesh)

colors = ['0.1', '0.3', '0.5', '0.7']
levels = [1, 2, 3, 4, 5]
plt.contourf(x, y, z, colors=colors, levels=levels)
```
